Находим частотность слов по дням с 1 по 21 апреля 1961

```python
dw = csvtools.DumpWrapper('/Users/mayakorotkaya/Desktop/prozhito-dump-1')
notes = dw.notes()
ALL = notes.sortInterval((1961, 4, 1), (1961, 4, 21))
```

```python
for el in ALL[0][1]:
    print(el.text)
```

```python
ALL[0]
```

```python
twoweeks = {}
for el in ALL:
    day = []
    for i in el[1]:
        day.append(i.text)
    twoweeks[el[0]] = day       
```

```python
twoweeks
```

```python
for key in twoweeks:
    mass = []
    for elem in twoweeks[key]:
        print(elem)
```

```python
twoweeks_lem = {}
for key in twoweeks:
    mass = []
    for elem in twoweeks[key]:
        elem = re.sub(punct,'',elem)
        elem = elem.split(' ')
        for z in elem:
            z = morph.parse(z)[0].normal_form
            if z!= ''  and z not in ru_stop:
                mass.append(z)
                c = Counter(mass)
                twoweeks_lem[key] = c.most_common()
```

```python
twoweeks_freq = {}
for key in twoweeks_lem:
    mass = []
    for elem in twoweeks_lem[key]:
        freq = elem[1]/len(twoweeks_lem[key])
        mass.append([elem[0], freq])
        if len(mass)<15:
            twoweeks_freq[key] = mass
        else:
            break
```

Посмотрим, какие слова частотны в конце 60-ых, а именно - около ввода войск в Чехословакию.

```python
dw = csvtools.DumpWrapper('/Users/mayakorotkaya/Desktop/prozhito-dump-1')
notes = dw.notes()
FIN = notes.sortInterval((1968, 8, 15), (1968, 8, 30))
```

```python
august = {}
for el in FIN:
    day_aug = []
    for i in el[1]:
        day_aug.append(i.text)
    august[el[0]] = day_aug   
```

```python
august_lem = {}
for key in august:
    massive = []
    for elem in august[key]:
        elem = re.sub(punct,'',elem)
        elem = elem.split(' ')
        for z in elem:
            z = morph.parse(z)[0].normal_form
            if z!= ''  and z not in ru_stop:
                massive.append(z)
                c = Counter(massive)
                august_lem[key] = c.most_common()
```

```python
august_freq = {}
for key in august_lem:
    mass = []
    for elem in august_lem[key]:
        freq = elem[1]/len(august_lem[key])
        mass.append([elem[0], freq])
        if len(mass)<15:
            august_freq[key] = mass
        else:
            break
```

```python
august_freq
```

```python
cheh = []
for key in august_freq:
    n = 0
    for el in august_freq[key]:
        if 'чехословакия' in el[0]:
            cheh.append(el[1])
        else:
            n+=1
    if n == 15:
        cheh.append(0)            
cheh
```

```python
dub = []
for key in august_freq:
    n = 0
    for el in august_freq[key]:
        if 'дубчек' in el[0]:
            dub.append(el[1])
        else:
            n+=1
    if n == 15:
        dub.append(0)            
dub
```

```python
voisko = []
for key in august_freq:
    n = 0
    for el in august_freq[key]:
        if 'войско' in el[0]:
            voisko.append(el[1])
        else:
            n+=1
    if n == 15:
        voisko.append(0)        
voisko
```

```python
kosh = []
for key in august_freq:
    n = 0
    for el in august_freq[key]:
        if 'кошица' in el[0]:
            kosh.append(el[1])
        else:
            n+=1
    if n == 15:
        kosh.append(0)        
kosh
```

```python
plt.plot(day_aug, cheh, color='red')
plt.plot(day_aug, dub, color='green')
plt.plot(day_aug, voisko, color='blue')
plt.plot(day_aug, kosh, color='black')
plt.xlabel('Day')
plt.ylabel('Frequency')

red_patch = mpatches.Patch(color='red', label='Чехословакия')
blue_patch = mpatches.Patch(color='blue', label='войско')
green_patch = mpatches.Patch(color='green', label='Дубчек')
black_patch = mpatches.Patch(color='black', label='Кошица')
plt.legend(handles=[red_patch, green_patch, blue_patch, black_patch]) 
plt.gcf().autofmt_xdate() 

plt.title('Ввод войск в Чехословакию')
plt.show()
```
